```python
from abc import ABC,abstractmethod

class Animal(ABC):
    @abstractmethod
    def move(self):
        pass
    def speak(self):
        pass

class Dog(Animal):
    def move(self):
        print('dog-walk')
    def speak(self):
        print('barking')

class Cat(Animal):
    def move(self):
        print('cat-walk')
    def speak(self):
        print('meow')

class Snake(Animal):
    def move(self):
        print('crawl')
    def speak(self):
        print('hisshisshisshisshisshiss')

himanshu=Dog()
himanshu.move()
himanshu.speak()

sheenu=Cat()
sheenu.move()
sheenu.speak()

sudhir=Snake()
sudhir.move()
sudhir.speak()
```

```python
from abc import ABC

class Vehicle(ABC):
    def __init__(self, brand, model, year, color="black"):
        self.brand = brand
        self.model = model
        self.year = year
        self.color = color

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

    @abstractmethod
    def honk(self):
        pass

    def display_info(self):
        print(f"{self.year} {self.color} {self.brand} {self.model}")

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

    def stop_engine(self):
        print("Car engine stopped")

    def honk(self):
        print("Car honking")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started")

    def stop_engine(self):
        print("Motorcycle engine stopped")

    def honk(self):
        print("Motorcycle honking")

# Usage
car = Car(brand="Toyota", model="Camry", year=2022, color="blue")
motorcycle = Motorcycle(brand="Harley-Davidson", model="Sportster", year=2021)

print("Car Information:")
car.display_info()
car.start_engine()
car.honk()
car.stop_engine()

print("\nMotorcycle Information:")
motorcycle.display_info()
motorcycle.start_engine()
motorcycle.honk()
motorcycle.stop_engine()
```

```python
from abc import ABC, abstractmethod

class ATMAccount(ABC):
    def __init__(self, balance=0):
        self.balance = balance

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def get_balance(self):
        pass

class SavingsAccount(ATMAccount):
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited RS.{amount:.2f} into Savings Account")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew RS.{amount:.2f} from Savings Account")
        else:
            print("Insufficient funds for withdrawal or invalid amount")

    def get_balance(self):
        return self.balance

class CurrentAccount(ATMAccount):
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited RS.{amount:.2f} into Current Account")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew RS.{amount:.2f} from Current Account")
        else:
            print("Insufficient funds for withdrawal or invalid amount")

    def get_balance(self):
        return self.balance

# Usage
savings_account = SavingsAccount(balance=1000)
Current_account = CurrentAccount(balance=500)

savings_account.deposit(200)
savings_account.withdraw(50)
print("Savings Account Balance:", savings_account.get_balance())

Current_account.deposit(150)
Current_account.withdraw(70)
print("Current Account Balance:", Current_account.get_balance())
```
